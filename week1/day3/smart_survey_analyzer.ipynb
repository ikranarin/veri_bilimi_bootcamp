```python
import random

scores = []

while len(scores) < 20:
    score = random.randint(0, 6) # 0 ile 6 arası rastgele sayı
    
    if score == 0:
        # 0 gelirse anketi hemen bitiriyoruz (break döngüyü tamamen sonlandırır)
        print("Anket kullanıcı tarafından sonlandırıldı (0 girildi).")
        break
        
    if score == 6:
        # 6 geçersiz bir puandır, yok sayıp döngünün başına dön
        continue
        
    # Geçerli puanları (1-5) listeye ekle
    scores.append(score)

print(f"Toplanan skorlar: {scores}")
```

```python
# len() fonksiyonu ile listedeki eleman sayısını bul
toplam_gecerli_cevap = len(scores)

print(f"Toplam geçerli cevap sayısı: {toplam_gecerli_cevap}")

# Neden while kullanıldı?
# Çünkü döngünün ne zaman biteceği sadece bir sayıya (20) bağlı değil, 
# aynı zamanda rastgele gelen 0 değerine de bağlıdır. Bu tip koşullu 
# durumlarda 'for' yerine 'while' kullanmak daha doğrudur.
```

```python
# sum() ile tüm puanları toplayıp eleman sayısına böl
if len(scores) > 0:
    ortalama = sum(scores) / len(scores)
    # round() kullanarak virgülden sonra 2 basamağa yuvarla
    ortalama_puan = round(ortalama, 2)
else:
    ortalama_puan = 0

print(f"Ortalama puan: {ortalama_puan}")
```

```python
# Veri seti boş değilse min ve max değerleri al
if scores:
    en_dusuk = min(scores)
    en_yuksek = max(scores)
else:
    en_dusuk = en_yuksek = 0

print(f"En düşük puan: {en_dusuk}")
print(f"En yüksek puan: {en_yuksek}")
```

```python
# Her kategori için sayaçları sıfırla
memnun_sayac = 0
kararsiz_sayac = 0
memnun_degil_sayac = 0

# Listenin içinde her bir skor için gez
for s in scores:
    if s >= 4:
        # 4 ve 5 memnun
        memnun_sayac += 1
    elif s == 3:
        # 3 kararsız
        kararsiz_sayac += 1
    else:
        # 1 ve 2 memnun değil
        memnun_degil_sayac += 1

print(f"Memnun: {memnun_sayac}")
print(f"Kararsız: {kararsiz_sayac}")
print(f"Memnun Değil: {memnun_degil_sayac}")
```

```python
# Arka arkaya kaç tane 1 puanı geldiğini takip eden bir değişken
arka_arkaya_bir = 0

for s in scores:
    if s == 1:
        arka_arkaya_bir += 1
    else:
        # Eğer puan 1 değilse sayaç sıfırlanmalı (çünkü 'arka arkaya' bozuldu)
        arka_arkaya_bir = 0
    
    # Koşul kontrolü: 3 adet oldu mu?
    if arka_arkaya_bir == 3:
        print("⚠️ Kritik memnuniyetsizlik tespit edildi!")
        break # Tespit edildiği anda döngüyü bitiriyoruz
```

```python
print("--- ANKET ANALİZ RAPORU ---")
print(f"Toplam geçerli cevap: {len(scores)}")
print(f"Ortalama puan: {ortalama_puan}")
print(f"Min puan: {en_dusuk}")
print(f"Max puan: {en_yuksek}")
print("-" * 25)
print(f"Memnun: {memnun_sayac}")
print(f"Kararsız: {kararsiz_sayac}")
print(f"Memnun değil: {memnun_degil_sayac}")
```
