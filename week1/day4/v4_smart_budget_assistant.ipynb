### Gerçek Hayatta Ne İşe Yarar?
Kişisel finans takibi, bütçe planlama ve harcama alışkanlıklarını analiz ederek tasarruf etme süreçlerini dijitalleştirmemizi sağlar.

### Kullanılan Python Konuları:
- Fonksiyonlar (def)
- Girdi/Çıktı (input, print)
- Sayısız Argüman (*args)
- Hata Yönetimi (try/except)
- Kapsam (Local vs Global Scope)
- Gömülü Metotlar (sum, len, round, max, min)

```python
def get_expenses():
    expenses = []
    try:
        count = int(input("Kaç adet harcama gireceksiniz?: "))
        for i in range(count):
            amount = float(input(f"{i+1}. harcama tutarını girin: "))
            expenses.append(amount)
    except ValueError:
        print("Hata: Lütfen geçerli bir sayı giriniz.")
    return expenses
```

```python
def calculate_summary(harcamalar_listesi):
    toplam = sum(harcamalar_listesi)
    ortalama = toplam / len(harcamalar_listesi) if len(harcamalar_listesi) > 0 else 0
    return toplam, round(ortalama, 2)
```

```python
def flexible_total(*args):
    print(f"İşlem yapılan harcama sayısı: {len(args)}")
    return sum(args)
```

```python
def find_extremes(harcamalar):
    return max(harcamalar), min(harcamalar)
```

```python
# 1. Adım: Verileri al
my_expenses = get_expenses()

if my_expenses:
    # 2. Adım: Fonksiyonları çağır
    toplam, ort = calculate_summary(my_expenses)
    en_yuksek, en_dusuk = find_extremes(my_expenses)
    
    # 3. Adım: Sonuçları yazdır
    print(f"\nToplam harcama: {toplam}{currency}")
    print(f"Ortalama harcama: {ort}{currency}")
    print(f"En yüksek harcama: {en_yuksek}{currency}")
    print(f"En düşük harcama: {en_dusuk}{currency}")
else:
    print("Harcama girilmedi veya bir hata oluştu.")
```
