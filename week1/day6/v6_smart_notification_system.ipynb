# PROJENİN AMACI
Farklı kanallardan (Email, SMS, Push) gelen bildirimleri, Polymorphism kullanarak tek bir arayüzden yönetmek.

### Polymorphism Nedir?
Farklı sınıfların aynı isimli metodun (send) farklı davranışlar sergilemesidir. Bu, sistemin esnekliğini artırır.

```python
from notifications import EmailNotification, SMSNotification, PushNotification
from utils import generate_message
```

```python
# Listemizi hazırlıyoruz
notifications = [
    EmailNotification(generate_message("Hoş geldin maili")),
    SMSNotification(generate_message("Doğrulama kodun: 1234")),
    PushNotification(generate_message("Yeni mesajın var!"))
]
```

```python
# Polymorphism'in gücü: Hangi tür bildirim olduğunu bilmemize gerek yok!
for notification in notifications:
    notification.send()
```

```python
# __str__ kullanımı
print(notifications[0])

# __len__ ile len() farkı
msg_len = len(notifications[0]) # __len__ metodumuzu çağırır
print(f"Mesaj uzunluğu: {msg_len}")
```
