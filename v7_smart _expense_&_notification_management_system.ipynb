## Projenin Amacı
Bu proje, kullanıcının finansal disiplin kazanmasını sağlamak amacıyla, harcamaların nesne tabanlı olarak modellendiği ve bütçe limitlerinin aşıldığı durumlarda çok biçimli (polymorphism) bildirim kanalları üzerinden uyarı üreten modüler bir finans yönetim sistemidir. Gerçek bir uygulama mimarisine sadık kalınarak hazırlanan bu sistemde; veri tutarlılığı, hata yönetimi ve modülerlik ilkeleri temel alınmış; harcama girişi, bütçe analizi ve otomatik bildirim süreçleri, birbirinden bağımsız çalışan ancak birbiriyle entegre modüllerle yönetilerek yazılım geliştirme disiplinine uygun, güvenli ve genişletilebilir bir altyapı oluşturulmuştur.

```python
from models import Budget
from notifications import EmailNotification, SMSNotification
from utils import get_user_expenses

budget = Budget(limit=5000)
```

```python
# Veri girişi ve sınıfa aktarım
user_data = get_user_expenses()
for exp in user_data:
    budget.add_expense(exp)
```

```python
total = budget.total_expense()
avg = budget.average_expense()
print(f"Toplam harcama: {total}TL")
print(f"Ortalama harcama: {avg:.2f}TL")
```

```python
notifications = []
if budget.is_limit_exceeded():
    print("Limit durumu: AŞILDI")
    notifications = [EmailNotification(), SMSNotification()]
else:
    print("Limit durumu: GÜVENLİ")

# Polymorphism ile aynı metodu çağırıyoruz
for note in notifications:
    print(f"-{note.send()}")
```
