Bu proje, gerçek hayattaki kütüphane sistemini (kitaplar, üyeler ve ödünç alma süreçleri) Nesne Yönelimli Programlama (OOP) prensipleriyle modellemektedir.

### OOP Neden Gereklidir?
Karmaşık sistemleri gerçek dünyadaki varlıklara benzer şekilde "nesne" olarak modellemek, kodun sürdürülebilirliğini, tekrar kullanılabilirliğini ve anlaşılabilirliğini artırır.

### Kullanılan OOP Kavramları:
- Class (Sınıf) Tanımı
- Attributes (Özellikler) ve Methods (Metotlar)
- Default (Varsayılan) Değerler
- Inheritance (Kalıtım)

```python
class Book:
    # __init__ metodu, nesne oluşturulduğunda çalışan yapıcı (constructor) metoddur
    def __init__(self, title, author, year, is_available=True):
        self.title = title
        self.author = author
        self.year = year
        self.is_available = is_available
        
    # Kitabın özet bilgilerini döndüren yardımcı metot
    def get_info(self):
        return f"{self.title} - {self.author} ({self.year})"
```

```python
class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        
    # Ödünç alma işlemi: Bir 'Book' nesnesini parametre olarak alır    
    def borrow_book(self, book):
        # Kitabın müsaitlik durumunu kontrol ediyoruz
        if book.is_available:
            book.is_available = False # Kitabı ödünç verdiğimiz için artık müsait değil
            print(f"{self.name} {book.title} kitabını ödünç aldı.")
        else:
            print("Kitap şu anda müsait değil.")
            
    # İade işlemi        
    def return_book(self, book):
        book.is_available = True # Kitabı geri aldığımızda tekrar müsait yapıyoruz
        print("Kitap iade edildi.")
```

```python
# DigitalBook sınıfı Book sınıfından miras alır (Inheritance)
class DigitalBook(Book):
    def __init__(self, title, author, year, file_size):
        # super() kullanarak üst sınıfın (Book) özelliklerini başlatıyoruz
        super().__init__(title, author, year)
        self.file_size = file_size # Dijital kitaba özel yeni özellik
        
    # Sadece dijital kitaplara özel metot    
    def download(self):
        print("Dosya indiriliyor...")
```

```python
# 1. En az 2 normal kitap oluştur
kitap1 = Book("Suç ve Ceza", "Dostoyevski", 1866)
kitap2 = Book("Nutuk", "Mustafa Kemal Atatürk", 1927)

# 2. 1 dijital kitap oluştur
dijital_kitap = DigitalBook("Python Rehberi", "Yazılım Ekibi", 2026, 15)

# 3. 1 üye oluştur
uye1 = Member("Ahmet", 101)

# 4. İşlemleri gerçekleştir
uye1.borrow_book(kitap1)  # Ödünç al
uye1.borrow_book(kitap1)  # Tekrar ödünç almayı dene (hata vermeli)
uye1.return_book(kitap1)  # Geri ver
dijital_kitap.download()  # Dijital metoda erişim
```
