```python
orders = [
    {
"customer":"Ahmet",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Zeynep",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]
```

```python
# Toplam sipariş sayısını len() fonksiyonu ile bul
toplam_siparis = len(orders)

# Ödenmiş siparişleri bool değerlerini tam sayıya çevirerek topla
# True değerleri 1, False değerleri 0 olarak döner
odenmis_sayisi = int(orders[0]["is_paid"]) + int(orders[1]["is_paid"]) + int(orders[2]["is_paid"]) + int(orders[3]["is_paid"])

# Toplamdan ödenmişleri çıkararak ödenmemiş sayısını bul
odenmemis_sayisi = toplam_siparis - odenmis_sayisi

print(f"Toplam sipariş: {toplam_siparis}")
print(f"Ödenmiş sipariş: {odenmis_sayisi}")
print(f"Ödenmemiş sipariş: {odenmemis_sayisi}")
```

```python
# Tüm müşteri isimlerini manuel olarak bir listeye topla
tum_musteriler = [orders[0]["customer"], orders[1]["customer"], orders[2]["customer"], orders[3]["customer"]]

# Tekrar eden isimleri temizlemek için listeyi set'e dönüştür
unique_musteriler = set(tum_musteriler)

print(f"Müşteri Listesi (Ham): {tum_musteriler}")
print(f"Benzersiz Müşteri Listesi: {unique_musteriler}")
```

```python
# Tüm ürün listelerini toplama operatörü ile birleştir
tum_urunler = orders[0]["items"] + orders[1]["items"] + orders[2]["items"] + orders[3]["items"]

# Unique ürünleri set kullanarak bul
unique_urunler = set(tum_urunler)

print(f"Tüm Siparişlerde Geçen Ürünler: {tum_urunler}")
print(f"Unique Ürünler: {unique_urunler}")
```

```python
# 1. Sipariş Kontrolü
if orders[0]["is_paid"]:
    print(f"{orders[0]['customer']} - Sipariş onaylandı")
else:
    print(f"{orders[0]['customer']} - Ödeme bekleniyor")

# 2. Sipariş Kontrolü
if orders[1]["is_paid"]:
    print(f"{orders[1]['customer']} - Sipariş onaylandı")
else:
    print(f"{orders[1]['customer']} - Ödeme bekleniyor")

# 3. Sipariş Kontrolü
if orders[2]["is_paid"]:
    print(f"{orders[2]['customer']} - Sipariş onaylandı")
else:
    print(f"{orders[2]['customer']} - Ödeme bekleniyor")
```

```python
# Müşteri adı ve tutarı içeren tuple'lar oluşturuyoruz
ozet_1 = (orders[0]["customer"], orders[0]["total"])
ozet_2 = (orders[1]["customer"], orders[1]["total"])
ozet_3 = (orders[2]["customer"], orders[2]["total"])

# Bu özetleri bir liste içinde topluyoruz
siparis_ozetleri = [ozet_1, ozet_2, ozet_3]

# TUPLE NEDEN DEĞİŞTİRİLEMEZ?
# Tuple'lar 'immutable' özelliğe sahiptir. Oluşturulduktan sonra içeriği değiştirilemez.
# Bu projede sipariş tutarı ve müşteri ismi gibi kritik verilerin kodun ilerleyen
# kısımlarında yanlışlıkla değiştirilmesini (manipüle edilmesini) önlemek için kullanılır.

print(f"Sipariş Özetleri: {siparis_ozetleri}")
```

```python
# Tüm analizlerin toplu özeti
print(f"Toplam sipariş:{toplam_siparis}")
print(f"Ödenmiş sipariş:{odenmis_sayisi}")
print(f"Ödenmemiş sipariş:{odenmemis_sayisi}")
print("-" * 20)
print(f"Unique ürünler: {unique_urunler}")
print(f"Sipariş özetleri: {siparis_ozetleri}")
```
